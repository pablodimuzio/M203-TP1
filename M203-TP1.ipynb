## Arc de cercle

### Exo1
Commençons par tracer le miroir. Testez la commande ci-dessous et modifiez-là pour obtenir un miroir demi-sphérique

#### Solution
On modifie le paramètre sector pour que l'arc de cercle aille de $-\frac{\pi}{2}$ à $\frac{\pi}{2}$. On en profite également pour mettre en place la variable 
`R` indiquant le rayon et qui sera utile plus tard.

```python
M=arc((0,0),1,1,sector=(-pi/2, pi/2))

R = 1

show(M,figsize=5)
```

## Lignes

### Exo 2 
On considère une ligne horizontale située à la distance h de l’axe optique, qui commence à l’abscisse $x = −1$ et qui s’arrête au point $A$ situé sur le miroir. Déterminer les coordonnées $x_A$ et $y_A$ du point $A$.

#### Solution:
On sait que le rayon coupe le cercle à $y=h$. Donc $\sin(\theta) = y$ et $\cos(\theta) = x$ avec $\theta \in [-\frac{\pi}{2},\frac{\pi}{2}]$. On déduis donc de la formule $\cos^2(\theta)+\sin^2(\theta) = 1$ que $x = \sqrt{1-h^2}$

```python
h = 0.5

xA = sqrt(1-h**2)
yA = h

A=point((xA, yA))

show(A+M)
```

### Exo 4 
Tracer deux lignes de ce type situées à des hauteurs différentes, $h_1 = 0.5$ et $h_2 = 0.8$, en vous inspirant du code déjà existant.

#### Solution:
On commence par définir $h_1$ et $h_2$. On en déduis donc $x_1$ et $x_2$ et on trace les rayons.

```python
h1 = 0.5
h2= 0.8

L1=line( [ (-1, h1), (sqrt(1-h1**2), h1) ])
L2=line( [ (-1, h2), (sqrt(1-h2**2), h2) ])
show(L1+L2+M,figsize=4)
```

## Tracé final

### Exo 12
Tracer, dans SAGE, le rayon réfléchi correspondant à $\frac{h}{R} = \frac{1}{2}$, en même temps que le rayon incident et le miroir.

```python
g=Graphics();

c= 0.5
h = R*c

xA = sqrt(1-h**2)
yA = h


L1=line( [ (-1, h), (xA, yA) ])
x=0.3
y=(x-xA)*(2*c)*(sqrt(1-c**2))/(1-2*c**2) + yA
L2 = line([(x,y), (xA, yA)])

g += L1 + L2
g+= M
show(g,figsize=4)
```

### Exo 13
Tracer l’ensemble des $8$ rayons réfléchis correspondant aux rayons incidents tracés plus haut.

```python
g=Graphics();
for n in range (1,9):
    h = n*0.1

    xA = sqrt(1-h**2)
    yA = h

    c= h/R

    L1=line( [ (-1, h), (xA, yA) ])
    y=-0.1
    x=(y-yA)/((2*c)*(sqrt(1-c**2))/(1-2*c**2))+xA
    L2 = line([(x,y), (xA, yA)])

    g += L1 + L2
g+= M
show(g,figsize=5)
```
