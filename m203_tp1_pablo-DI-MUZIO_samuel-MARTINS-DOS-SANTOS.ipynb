### Exo 2 
On considère une ligne horizontale située à la distance h de l’axe optique, qui commence à l’abscisse $x = −1$ et qui s’arrête au point $A$ situé sur le miroir. Déterminer les coordonnées $x_A$ et $y_A$ du point $A$.

#### Solution:
On sait que le rayon coupe le cercle à $y=h$. Donc $\sin(\theta) = y$ et $\cos(\theta) = x$ avec $\theta \in [-\frac{\pi}{2},\frac{\pi}{2}]$. On déduis donc de la formule $\cos^2(\theta)+\sin^2(\theta) = 1$ que $x = \sqrt{1-h^2}$

```sage
h = 0.2

xA = sqrt(1-h**2)
yA = h

A=point((xA, yA))

show(A+M)
```

## Un peu de trigonométrie
Considérons un rayon incident distant de $h$ de l’axe optique. Il fait un angle $\theta$ avec la normale au miroir, au point $A$ où il l’atteint.

### Exo 8 
Exprimer $\theta$ en fonction de $h$ et de $R$, le rayon de courbure du miroir. Montrer que 

$$
\sin(\theta) = \frac{h}{R} \quad\quad \cos(\theta) = \sqrt{1 - \frac{h^2}{R^2}}
$$

On sait grace aux propriétés trigonométriques que $\sin(\theta) = \frac{h}{R}$. On utilise la même déduction qu'à l'exercice 2 pour trouver le résultat $\cos(\theta) = \sqrt{1-\sin^2(\theta)} \implies \cos(\theta) = \sqrt{1-\frac{h^2}{R^2}}$

### Exo 9 
Le rayon réfléchi fait un angle $2\theta$ avec l’horizontale. Montrer que l’équation du rayon réfléchi s’écrit

$$
y−y_A =(x−x_A)\tan(2\theta)
$$

#### Solution:
On a $\frac{y-y_A}{x-x_A} = P$ où P est la pente du rayon réfléchi. On sait aussi avec les propriétés trigonométriques que $\tan(2\theta) = \frac{h}{x} = P$. Ainsi, on a $\frac{y-y_A}{x-x_A} = \tan(2\theta) \implies y-y_A = (x-x_A) \tan(2\theta)$

### Exo 10 
Aller sur le site <a href="www.wolframalpha.com"><b>Wolfram Alpha</b></a> et entrer $\tan(2x)$. Montrer le premier résultat indiqué

#### Solution:

$$ \tan(2x) = \frac{2 \sin(x) \cos(x)}{ \cos^2(x) − \sin^2(x)} $$

### Exo 11 
Reprendre Exo 9 et écrire l’équation du rayon réfléchi en fonction de $x_A$, $y_A$, $h$ et $R$.

#### Solution:

$$y-y_A = (x-x_A) \frac{ 2  \frac{h}{R} \sqrt{1- \frac{h^2}{R^2}}}{1-2 \frac{h^2}{R^2}}$$

Après différents tests, nous nous sommes rendu comte qu'il était plus facile de fixer `y` et de calculer `x`. On en arrive donc à l'équation suivante:

$$x = \frac{(y-y_A)(1-2\frac{h^2}{R^2})}{2\frac{h}{R}\sqrt{1-\frac{h^2}{R^2}}} + x_A$$

## Tracé final

### Exo 12
Tracer, dans SAGE, le rayon réfléchi correspondant à $\frac{h}{R} = \frac{1}{2}$, en même temps que le rayon incident et le miroir.

#### Solution:
On pose `c=h/R` pour plus de facilités dans la programmation. On calcule l'équation du rayon incident et l'équation du rayon réfléchi avec les équations définies précedemment. On pose `y=-0.1` afin de voir l'intersection entre les rayons et l'axe optique 

```sage
g=Graphics();

c= 0.5
h = R*c

xA = sqrt(1-h**2)
yA = h


L1=line( [ (-1, h), (xA, yA) ])
y=-0.1
x=(y-yA)/((2*c)*(sqrt(1-c**2))/(1-2*c**2))+xA
L2 = line([(x,y), (xA, yA)])

g += L1 + L2
g+= M
show(g,figsize=4)
```

### Exo 13
Tracer l’ensemble des $8$ rayons réfléchis correspondant aux rayons incidents tracés plus haut.

#### Solution:
On utilise une boucle `for` afin de générer tous les rayons incidents, puis les rayons réfléchis. On plot le tout et on optient le graph qui suit.

```sage
g=Graphics();
for n in range (1,9):
    h = n*0.1

    xA = sqrt(1-h**2)
    yA = h

    c= h/R

    L1=line( [ (-1, h), (xA, yA) ])
    y=-0.1
    x=(y-yA)/((2*c)*(sqrt(1-c**2))/(1-2*c**2))+xA
    L2 = line([(x,y), (xA, yA)])

    g += L1 + L2
g+= M
show(g,figsize=12)
```

### Exo 15
Reprenez l’exercice en considérant la traversée d’une lentille plan-convexe d’indice de réfraction $n$. On pourra comparer la position du point de convergence des rayons paraxiaux avec ce qu’indique la formule dite des opticiens dans ce cas particulier

$$f=\frac{R}{n−1}$$

#### Construction de la lentille
On commence par construire une lentille plan-convexe à rayon de paramètre `R`

```sage
R=1

OffsetX, OffsetY = 1,0

Lens=arc((OffsetX,OffsetY),R,R,sector=(-pi/2, -3*pi/2), color="red")

Lens+= line([(OffsetX,R+OffsetY),(OffsetX-R, OffsetY+R)], color="red")
Lens+= line([(OffsetX,OffsetY-R), (OffsetX-R, OffsetY-R)], color="red")
Lens+= line([(OffsetX-R,OffsetY+R), (OffsetX-R,OffsetY-R)], color="red")

show(Lens,figsize=5)
```

#### Construction des rayons incidents

```sage
L1=line( [ (-2+OffsetX, h+OffsetY), (OffsetX-sqrt(1-(h+OffsetY)**2), h+OffsetY) ])
show(L1+Lens,figsize=4)
```

```sage
g=Graphics();
for n in range (1,9):
    h = n*0.1
    L1=line( [ (-2+OffsetX, h+OffsetY), (OffsetX-sqrt(1-(h+OffsetY)**2), h+OffsetY) ])
    g += L1
g+= Lens
show(g,figsize=4)
```

#### Construction des rayons réfractés

On fixe l'indice de réfraction de la lentille à une valeur arbitraire

```sage
INDICE_OPT = 1.5
```

On connait grace à la loi de Snell-Descartes la relation entre $n, i$ et $r$ avec $i$ et $r$ les angles par rapport à la normale des rayons réfractés:
$$\sin(i) = n' \sin(r)$$

On en déduit l'angle du rayon réfracté: $r = \arcsin(\frac{\sin(i)}{n'})$

On sait que $\tan(r) = x$ avec $x$ la distance entre le projeté de $A$, le point d'incidence sur l'axe $X$ et le centre du cercle. On a aussi $\tan(r) = \frac{x}{h}$ 

On en déduis que $x = h\tan(r)$

Sait aussi que $\sin(i) = \frac{h}{R}$

$$x = h\tan(\arcsin(n'\frac{h}{R}))

On implémente ceci en python:

```sage
g=Graphics();
for n in range (1,9):
    h = n*0.1

    yA = h
    xA = -sqrt(1-(h+OffsetY)**2)
    L1=line( [ (-2+OffsetX, yA+OffsetY), (xA+OffsetX, yA+OffsetY) ], color="green")

    x = h*tan(arcsin(INDICE_OPT*h/R))

    L2 = line( [ (xA+OffsetX, yA+OffsetY), (x, 0) ], color="blue")
    
    g+=L1 +L2


g+= Lens
show(g,figsize=4)
```
